#Matplotlib

   - useful for making plots

```python
#importing matplotlib Library

import matplotlib.pyplot as plt
```

```python
# import numpy to get data for our plots
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
# sin wave
plt.plot(x,y)
plt.show()
```

```python
# cos wave
plt.plot(x,z)
plt.show()
```

```python
# adding title x-axis and y-axis labels
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('sin-wave')
plt.show()
```

```python
# parabola

x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'g+')
plt.show()
```

```python
plt.plot(x,y,'y.')
plt.show()
```
