#Matplotlib

   - useful for making plots

```python
#importing matplotlib Library

import matplotlib.pyplot as plt
```

```python
# import numpy to get data for our plots
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
# sin wave
plt.plot(x,y)
plt.show()
```

```python
# cos wave
plt.plot(x,z)
plt.show()
```

```python
# adding title x-axis and y-axis labels
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('sin-wave')
plt.show()
```

```python
# parabola

x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'g+')
plt.show()
```

```python
plt.plot(x,y,'y.')
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x, np.sin(x), '--')
plt.plot(x, np.cos(x),'r-')
plt.show
```

Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English', 'French', 'German','Spanish',"Latin"]
students = [23,17,35, 76,80]
ax.bar(languages, students)
plt.xlabel('Languages')
plt.ylabel('Students')
plt.title('Students per Language')
plt.show
```

Pie Chart

```python
fig1 = plt.figure()
ax1 = fig1.add_axes([0,0,1,1])
languages = ['English', 'French', 'German','Spanish',"Latin"]
people = [23,17,35, 76,80]
ax1.pie(people, labels= languages, autopct = '%1.1f%%')
plt.show()
```

Scatter Plot


```python
x= np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color = 'g')
ax.scatter(x,z,color='b')
plt.show
```

3D Scatter plot


```python
fig3 = plt.figure()
ax = plt.axes(projection = '3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter3D(x,y,z,c=z, cmap = 'Greens')
plt.show()
```
