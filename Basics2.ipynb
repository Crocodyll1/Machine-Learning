```python
#Arithmetic Operator
num_1 = 20
num_2 = 10

#aaddition
sum = num_1 + num_2
print("Sum is : ", sum)

#substraction
diff = num_1 - num_2
print("Difference is : ", diff)

#multiplication
pro= num_1 * num_2
print("Product is : ", pro)

#Division
div = num_1 / num_2
print("Division is : ", div)

#exponent
exp = num_1 ** num_2
print("Exponent is : ", exp)

#modulud
mod = num_1 % num_2
print("Modulus is : ", mod)




```
