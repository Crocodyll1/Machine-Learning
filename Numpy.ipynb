List vs Nimpy - Time taken

```python
from time import process_time
```

Time taken by a list

```python
python_list =[i for i in range(10000)]

start_time = process_time()

python_list = [i+5 for i in python_list]

end_time = process_time()

print(end_time - start_time)
```

```python
np_array = np.array([i for i in range(10000)])

start_time = process_time()

np_array += 5

end_time = process_time()

print(end_time - start_time)
```

Mathematical oprerations on a np array

```python
list1 = [1,2,3,4,5]
list2 = [6,7,8,9,10]

print(list1 + list2) # concatenate or joins two list

```

```python
k =  np.random.randint(0,10,(3,3))
l =  np.random.randint(10,20,(3,3))

print(k)
print(l)

```

```python
print(k + l)
print(k - l)
print(k * l)
print(k / l)
#
```

```python
k =  np.random.randint(0,10(3,3))
l =  np.random.randint(10,20(3,3))

print(k)
print(l)
```

```python
print(np.add(k,l))
print(np.subtract(k,l))
print(np.multiply(k,l))
print(np.divide(k,l))
#
```

Array Manipulation

```python
array = np.random.randint(0,10,(2,3))
print(array)
print(array.shape)
```

```python
# transpose
trans = np.transpose(array)
print(trans)
print(trans.shape)
```

```python
trans2 = array.T
print(trans2)
print(trans2.shape)
```

```python
# Reshaping a array
a = np.random.randint(0,20,(2,3))

print(a)
print(a.shape)

b = a.reshape(3,2)
print(b)
print(b.shape)
```
